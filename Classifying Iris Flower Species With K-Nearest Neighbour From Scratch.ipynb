To study more about partial corrleation, visit https://en.wikipedia.org/wiki/Partial_correlation.

```r
# Compute partial correlation between sepal length and width by controlling R.V. petal length and width
cov_iris <- cov(iris[,-5])
prtl_S <- cov_iris[1:2,1:2] - cov_iris[1:2,3:4] %*% solve(cov_iris[3:4,3:4]) %*% cov_iris[3:4,1:2]
prtl_cor <- prtl_S[1,2]/sqrt(prtl_S[1,1] * prtl_S[2,2])
sprintf("The partial correlation is %.3f.", prtl_cor)
```

The correlation coeifficient increases from -0.118 to 0.629. In my opinion, it is resonable to expect that the longer sepal length the Iris flower has, the longer the sepal width it has. Hence, the result looks much acceptable after controlling the effect on petal length and petal width.

## Creating Custom Function for K Nearest Neighbours Classifier

The `kNN` function can be used to build the classifier and make prediction.

The `kNN` function takes the following argument:
* **test** - testing vector
* **Train** - matrix or data frame of trainning data
* **Class** - factor/category for each row of trainning data
* **k** - number of nearest neighbours the model will consider

```r
# Create kNN function
kNN <- function(test, Train, Class, k){
    test <- as.matrix(test); Train <- as.matrix(Train); Class <- as.matrix(Class)
    distance <- NULL
    for(i in 1:nrow(Train)){
        distance[i] <- dist(rbind(test, Train[i,]))
    }
    temp <- cbind(distance, Class)
    ClassSort <- temp[order(distance),]
    class <- names(which.max(table(ClassSort[1:k,2])))
    return(class)
}
```

## Model Trainning

Now I'll normalise the dataset before trainning the model, and then I'll shuffle and split them into trainning and testing dataset.

```r
# Normalise dataset
iris2 <- as.data.frame(lapply(iris[,1:4], scale))
iris2 <- cbind(iris2, iris$Species)

# Shuffle dataset
rand <- sample(nrow(iris2))
iris3 <- iris2[rand,]

# Split dataset
split_pt <- round(nrow(iris3)*0.25,0)
Test <- iris3[1:split_pt,1:4]
Train <- iris3[(split_pt+1):nrow(iris3),1:4]
Class <- iris3[(split_pt+1):nrow(iris3),5]
```

Now use the custom function `kNN` to build the model and make prediction on the Iris flower dataset.

```r
# Build model and make prediction
result <- NA
for(i in 1:nrow(Test)){
    result[i] <- kNN(Test[i,], Train, Class, k=21)
}
```

The `result` variable saved the predicted Iris species of testing dataset.

## Evaluating Model Performance

Now I'll evaluate the model performance using contingency table.

```r
# Evaluate performance
real <- iris3[1:split_pt,5]
CrossTable(result, real, prop.chisq=F, prop.t=F, dnn=c("predict","actual"))
```

There are 4 out of 38 Iris species cannot be classifed to the correct category. However, in the absence of sufficient samples, the model can still correctly classify the Iris flower species with about 90% accuracy. In my opinion, the model has performed very well.

To enhance the accuracy of the classifier, we can increase the sample size of trainning dataset or adjust the value of k used in `kNN` function.
